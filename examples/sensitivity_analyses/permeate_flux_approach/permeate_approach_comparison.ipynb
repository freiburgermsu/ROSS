# Linear permeate vs. linear CF

## Linear CF - Brine

```python
# run the ROSS model
# %run ../../../rosspy/core.py
# ross = ROSSPkg()
import rosspy
ross = rosspy.ROSSPkg(verbose = False)


# general parameters
database_selection = 'pitzer'
simulation_title = 'test'
water_selection = 'red_sea'
simulation = 'brine'
simulation_perspective = 'all_time'
simulation_time = 75
final_cf = 1.117096

# create the input file
ross.define_general(database_selection, simulation)
ross.transport(simulation_time, simulation_perspective)
ross.reaction(final_cf)
ross.solutions(water_selection)
ross.equilibrium_phases()
ross.selected_output()
ross.export()

# execute the input file
raw_data = ross.execute()
brine_data = ross.process_selected_output()
display(brine_data)
```
