# Scaling with all of the databases

```python
%matplotlib inline
%run ../../../../rosspy/ro.py
# import rosspy

# simulate the databases
ross = ROSSPkg('pitzer')
for db in ross.databases:
    print('='*30, f'\n\n{db} database \n', '='*(len(db)+10))
    export_name = f'{db} database'

    # define the simulation
    ross = ROSSPkg(db, export_content = True)
    ross.reactive_transport(simulation_time = 200)
    ross.feed_geochemistry('red_sea')
    
    # execute and process the input file
    try:
        ross.execute(export_name, figure_title = f'Scaling in the Red Sea via the {db} database', export_format="eps")
        display(ross.processed_data)
    except:
        print(f'\n\n--> ERROR: The simulation with the {db} database failed to converge\n\n')
    
```
