# Linear permeate vs. linear CF

## Linear CF - Brine

```python
%run ../../../../rosspy/ro.py
# import rosspy


# general parameters
database_selection = 'pitzer'
water_selection = 'red_sea'
simulation = 'brine'
simulation_perspective = 'all_time'
simulation_time = 75
final_cf = 1.117096

# execute the simulation
ross = ROSSPkg(database_selection, simulation, printing = False)
ross.reactive_transport(simulation_time, simulation_perspective, final_cf)
ross.feed_geochemistry(water_selection)
ross.execute(export_format="eps")
display(ross.processed_data)
```
