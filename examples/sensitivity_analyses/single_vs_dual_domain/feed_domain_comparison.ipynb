# Single versus Dual domain

## Single domain

### Brine

```python
%run ../../../rosspy/core.py
# import rosspy

# general parameters
simulation_name = 'single_domain-brine'
database_selection = 'pitzer'
water_selection = 'red_sea'
simulation = 'brine'
simulation_perspective = 'all_time'
simulation_time = 75

# create the input file
ross = ROSSPkg(database_selection, simulation)
ross.reactive_transport(simulation_time, simulation_perspective)
ross.feed_geochemistry(water_selection)
ross.execute(simulation_name)

# execute the input file
display(ross.processed_data)
```
