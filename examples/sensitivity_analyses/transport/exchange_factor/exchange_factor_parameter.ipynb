# exchange_factor parameter of the Dual domain 

## Large value

```python
# run the ROSS model
# %run ../../../rosspy/ro.py
# ross = ROSSPkg()
import rosspy
ross = rosspy.ROSSPkg(verbose = False)

# general parameters
simulation_name = 'large_EF-mobile'
database_selection = 'pitzer'
simulation_title = 'test'
water_selection = 'red_sea'
simulation = 'brine'
domain = 'dual'
domain_phase = 'mobile'
simulation_perspective = 'all_distance'
simulation_time = 75
exchange_factor = 1e10

# create the input file
ross.define_general(database_selection, simulation, domain, domain_phase)
ross.transport(simulation_time, simulation_perspective, exchange_factor = exchange_factor)
ross.reaction()
ross.solutions(water_selection)
ross.equilibrium_phases()
ross.selected_output()
ross.export(simulation_name)

# execute the input file
raw_data = ross.execute()
brine_data = ross.process_selected_output()
```
