# Evaluation of various feed water sources

```python
%run ../../../rosspy/core.py

# initial parameters
database_selection = 'pitzer'
simulation_time = 200

# create the input file
ross = ROSSPkg(database_selection)
for feed in ross.feed_sources:
    print(feed,'\n', '='*len(feed))
    
    ross = ROSSPkg(database_selection)
    ross.reactive_transport(simulation_time)
    ross.feed_geochemistry(feed)
    ross.execute(f'{feed} scaling', plot_title = f'Scaling from the {feed}')
```
