# Red Sea scaling through BW30-400

## Evaluate scaling along the module at the final time

```python
# run the ROSS model
%run ../../../rosspy/ro.py
# import rosspy

# initial parameters
database_selection = 'pitzer'
simulation_title = 'test'
water_selection = 'red_sea'
simulation_perspective = 'all_distance'
simulation_time = 200
simulation_name = 'normal'

# create the input file
ross = ROSSPkg(database_selection, printing = True, export_content = True)
ross.reactive_transport(simulation_time, simulation_perspective)
ross.feed_geochemistry(water_selection)
ross.execute(simulation_name)

# execute and process the input file
display(ross.processed_data)
```

## Evaluate scaling over a custom, high permeate, module at the final time

```python
# run the ROSS model
%run ../../../rosspy/ro.py
# import rosspy

# initial parameters
database_selection = 'pitzer'
simulation_title = 'test'
water_selection = 'red_sea'
simulation_time = 200
simulation_perspective = 'all_distance'
simulation_name = 'high-permeate'

# custom module parameterization
module_characteristics = {
    'permeate_flow_m3_per_hour': {
        'value':12.5
    }
}

# create the input file
ross = ROSSPkg(database_selection, printing = True)
ross.reactive_transport(simulation_time, simulation_perspective, None, module_characteristics)
ross.feed_geochemistry(water_selection)
ross.execute(simulation_name)

# execute and process the input file
display(ross.processed_data)
```
