# Define and execute the simulation

```python
# execute the ROSSpy simulation
# %run ../../../../rosspy/ro.py
from rosspy import ROSSPkg
import pandas
# pandas.set_option('max_columns',None)

# initial parameters
database_selection = 'pitzer'
simulation_perspective = 'all_distance'
simulation_time = 200

# Define the feed conditions
water_selection = 'Spacer_filled_channels'
water_characteristics = {
    'element': {
        'Ca': {
            'concentration (ppm)': 800
            }, 
        'S(6)': {
            'concentration (ppm)': 1975
        }
    },
    'pH': {
        'value': 6.4
    },
}

# create the input file
ross = ROSSPkg(database_selection, export_content = False)
ross.reactive_transport(simulation_time, simulation_perspective)
ross.feed_geochemistry(water_selection, water_characteristics)
ross.execute()

# execute and process the input file
display(ross.processed_data)
display(ross.selected_output)
```
