# Evaluation of various feed water sources

```python
from pandas import set_option
%run ../../../../rosspy/ro.py

# initial parameters
database_selection = 'pitzer'
simulation_time = 200

set_option('display.max_columns', None)

# create the input file
ross = ROSSPkg(database_selection)
for feed in ross.feed_sources:
    print(feed,'\n', '='*len(feed))
    
    ross = ROSSPkg(database_selection, export_content = False)
    ross.reactive_transport(simulation_time)
    ross.feed_geochemistry(feed)
    ross.execute(f'{feed} scaling', figure_title = f'Scaling from the {feed}', export_format="eps")
```
