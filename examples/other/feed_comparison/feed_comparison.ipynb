# Evaluation of various feed water sources

```python
# run the ROSS model
from glob import glob
import json
import os
%run ../../../rosspy/core.py
ross = ROSSPkg()
# import rosspy
# ross = rosspy.ROSSPkg(verbose = False)

# initial parameters
quantity_of_modules = 1
database_selection = 'pitzer'
simulation_title = 'test'
simulation_perspective = 'all_distance'
simulation_time = 200

# create the input file
for water_body in ross.feed_sources:
    print(water_body,'\n', '='*len(water_body))
    
    ross = ROSSPkg(verbose = False)
    ross.define_general(database_selection, quantity_of_modules = quantity_of_modules)
    ross.transport(simulation_time, simulation_perspective, )
    ross.reaction()
    ross.solutions(water_body)
    ross.equilibrium_phases()
    ross.selected_output()
    ross.export()

    # execute and process the input file
    raw_data = ross.execute()
    distance_data = ross.process_selected_output()
    display(distance_data)
```
