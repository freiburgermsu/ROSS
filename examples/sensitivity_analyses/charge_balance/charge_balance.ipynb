# pH charge balance

```python
%run ../../../rosspy/core.py
ross = ROSSPkg()

# define the simulation parameters
parameterized_ph_charge = True
database_selection = 'pitzer'
simulation = 'brine'
water_selection = 'red_sea'
simulation_perspective = 'all_time'
simulation_time = 275
simulation_name = 'balanced_charge'

# execute the simulation
ross.define_general(database_selection, simulation)
ross.transport(simulation_time, simulation_perspective, )
ross.reaction()
ross.solutions(water_selection, parameterized_ph_charge = parameterized_ph_charge)
ross.equilibrium_phases()
ross.selected_output()
ross.export(simulation_name)

# execute and process the input file
raw_balanced_data = ross.execute()
processed_balanced_data = ross.process_selected_output()
```
