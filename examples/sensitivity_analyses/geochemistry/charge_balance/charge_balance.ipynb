# pH charge balance

```python
%run ../../../rosspy/ro.py
# import rosspy

# define the simulation parameters
parameterized_ph_charge = True
database_selection = 'pitzer'
simulation = 'brine'
water_selection = 'red_sea'
simulation_time = 275
simulation_name = 'balanced_charge'

# execute the simulation
ross = ROSSPkg(database_selection, simulation, jupyter = True)
ross.reactive_transport(simulation_time)
ross.feed_geochemistry(water_selection, parameterized_ph_charge = parameterized_ph_charge)
ross.execute(simulation_name, plot_title = f'Brine with a {simulation_name}')
```
